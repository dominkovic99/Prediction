```python
import pandas as pd
podaci = pd.read_csv('Datasets\\sezone.csv')
podaci.head()
```

```python
podaci["HomeTeam_cat"] = podaci["HomeTeam"].astype('category').cat.codes
podaci["AwayTeam_cat"] = podaci["AwayTeam"].astype('category').cat.codes
podaci["FTR_cat"] = podaci["FTR"].astype('category').cat.codes
podaci["Referee_cat"] = podaci["Referee"].astype('category').cat.codes
podaci["HM1_cat"] = podaci["HM1"].astype('category').cat.codes
podaci["HM2_cat"] = podaci["HM2"].astype('category').cat.codes
podaci["HM3_cat"] = podaci["HM3"].astype('category').cat.codes
podaci["HM4_cat"] = podaci["HM4"].astype('category').cat.codes
podaci["HM5_cat"] = podaci["HM5"].astype('category').cat.codes
podaci["AM1_cat"] = podaci["AM1"].astype('category').cat.codes
podaci["AM2_cat"] = podaci["AM2"].astype('category').cat.codes
podaci["AM3_cat"] = podaci["AM3"].astype('category').cat.codes
podaci["AM4_cat"] = podaci["AM4"].astype('category').cat.codes
podaci["AM5_cat"] = podaci["AM5"].astype('category').cat.codes
podaci.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(podaci[['HomeTeam_cat','AwayTeam_cat', 'B365H', 'B365D', 'B365A', 'HTGS', 'ATGS', 'HTGC', 'ATGC', 'HTP', 'ATP','HM1_cat','HM2_cat','HM3_cat','HM4_cat','AM1_cat','AM2_cat','AM3_cat','AM4_cat']], 
                                                    podaci[['FTR_cat']], test_size=0.25, random_state=2)
```

```python
X_train = X_train.astype(float)
X_test = X_test.astype(float)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier


lr = LogisticRegression(random_state=2)
rf = RandomForestClassifier(random_state=2)
gbc = GradientBoostingClassifier(random_state=1)
dtc = DecisionTreeClassifier(random_state=3)
knn = KNeighborsClassifier(n_neighbors=3)

lr.fit(X_train_scaled, y_train)
rf.fit(X_train_scaled, y_train)
gbc.fit(X_train_scaled, y_train)
dtc.fit(X_train_scaled, y_train)
knn.fit(X_train_scaled, y_train)
```

```python
print('Train Score')
print('LR:' + str(lr.score(X_train_scaled, y_train)))
print('RF:' + str(rf.score(X_train_scaled, y_train)))
print('GBC:' + str(gbc.score(X_train_scaled, y_train)))
print('DTC:' + str(dtc.score(X_train_scaled, y_train)))
print('KNN:' + str(knn.score(X_train_scaled, y_train)))
print('--------------------------------')
print('Test Score')
print('LR:' + str(lr.score(X_test_scaled, y_test)))
print('RF:' + str(rf.score(X_test_scaled, y_test)))
print('GBC:' + str(gbc.score(X_test_scaled, y_test)))
print('DTC:' + str(dtc.score(X_test_scaled, y_test)))
print('KNN:' + str(knn.score(X_test_scaled, y_test)))
```
